```python
import pandas as pd 

df = pd.read_csv("general_data.csv")
```

```python
df.head()
```

```python
df.Attrition = df.Attrition.map({"Yes" : 1, "No" : 0}) 
df.Gender = df.Gender.map({'Male' : 1, 'Female' : 0})
```

```python
df1 = df.dropna()
```

__Correlation Matrix__

```python
df1.corr() 
```

__Correlation coefficient between Attrition and Age__

```python
from scipy.stats import pearsonr 
stats,p = pearsonr(df1.Attrition,df1.Age)

print('the value of correlation co-efficient :',stats)
if stats < 0:
    print("Hence they are negatively correlated")
elif stats > 0:
    print('Hence they are positively correlated')
elif stats == 0:
    print('Hence they are not correlated')
```
